```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("ColorHistogram.asc", header=None, sep=' ')
data = np.array(df)[:, 1:]
data.shape
```

```python
df.iloc[:, 1:]
```

```python
# 计算协方差矩阵
covMat = np.cov(data, rowvar=False)
covMat.shape
```

```python
# 计算特征值和特征向量
eigVals, eigVects = np.linalg.eig(np.mat(covMat))
eigVals.shape, eigVects.shape
```

```python
eigVals
```

```python
# 对特征值从小到大排序
eigValIndice = np.argsort(eigVals)
eigValIndice
```

```python
top = 3
# 最大top个特征值的下标
n_eigValIndice = eigValIndice[-1: -(top+1): -1]
n_eigValIndice
```

```python
n_eigVect = eigVects[:, n_eigValIndice]
```

```python
# 低维特征空间的数据
lowDDataMat = np.matrix(data) * n_eigVect
lowDDataMat.shape
```

```python
import matplotlib.pyplot as plt
import cv2

img_path = "6e2efc7cc92c7ee4.jpg"
image=cv2.imread(img_path,0)

plt.imshow(image,cmap='gray')
plt.show()
image.shape
```

```python
fl = image.flatten()
x = list(range(256))
count = [0] * 256
for i in fl:
    count[i] += 1
plt.bar(x, count, fc = 'skyblue') 
plt.show()
```

```python
from sklearn.cluster import KMeans

# 质心数
K_nums = 4

kmeans=KMeans(n_clusters=K_nums, n_init=10)
img_data=image.reshape((-1,1))
kmeans.fit(img_data) 
centroids=kmeans.cluster_centers_.squeeze() # 每一个类别的质心
labels=kmeans.labels_ # 每一个类别的标记
centroids
```

```python
cts = list(np.around(centroids))
count_ = [count[int(i)] for i in cts]
plt.bar(x, count, fc = 'skyblue')
plt.bar(cts, count_, fc = 'r')
```

```python
img2 = np.choose(labels,centroids).reshape(image.shape)
```

```python
plt.imshow(img2, cmap='gray')
plt.show()
```
